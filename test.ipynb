```python

key = 'a82d1438bbbc4ac5865ecde123d56861'
q = 'taylor swift'
fromd = '2023-11-05'
tod = '2023-11-05'
url = f'https://newsapi.org/v2/everything?q={q}&from={fromd}&to={tod}&sortBy=popularity&apiKey={key}'
response = requests.get(url)
data = response.json()
```

```python
news_data = []

# Iterate through each article in the 'articles' list
for article in data['articles']:
    source = article['source']['name']
    author = article['author']
    title = article['title']
    description = article['description']
    url = article['url']
    urlToImage = article['urlToImage']
    publishedAt = article['publishedAt']
    content = article['content']
    
    # Split the 'publishedAt' into date and time
    published_date, published_time = publishedAt.split('T')
    
    # Append the extracted data to the list
    news_data.append([source, author, title, description, url, urlToImage, published_date, published_time, content])

# Create a DataFrame from the extracted data
df = pd.DataFrame(news_data, columns=['Source', 'Author', 'Title', 'Description', 'URL', 'URLToImage', 'PublishedDate', 'PublishedTime', 'Content'])
df['nfl_presence'] = df['Content'].apply(lambda x: 'Yes' if 'Travis' in x else 'No')

# Add a column for 'sex_or_ex_presence'
df['sex_or_ex_presence'] = df['Description'].apply(lambda x: 'Yes' if 'sex' in x.lower() or 'ex' in x.lower() else 'No')
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_').str.strip()
df
```
